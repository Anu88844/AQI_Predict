```python
y=AQI.iloc[:,6]
y
```

```python
x=AQI[['co','no','no2','o3','so2','pm2_5','nh3']]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,
                                                   y,
                                                   test_size=0.20,
                                                   random_state=0)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
AQI_Predict=LinearRegression()
```

```python
AQI_Predict.fit(x_train,y_train)
```

```python
m = AQI_Predict. coef_
c = AQI_Predict.intercept_
m,c
```

```python
c
```

```python
y_predict_train = AQI_Predict.predict(x_train)
y_predict_train
```

```python
y_predict_test = AQI_Predict.predict(x_test)
y_predict_test
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test,y_predict_test)
```

```python
mean_absolute_error(y_train,y_predict_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_predict_test)
```

```python
mean_squared_error(y_train,y_predict_train)
```

```python
accuracy=r2_score(y_test,y_predict_test)
```

```python
accuracy
```

```python
print("accuracy is",accuracy*100)
```

```python
r2score_train= r2_score(y_train, y_predict_train)
r2score_train
```

```python
plt.style.use("default")

plt.figure(figsize=(4, 4))
sns.regplot(x=y_test, y=y_predict_test, ci=None, color="red")
```

```python
plt.figure(figsize=(4,4))
plt.style.use("ggplot")

plt.scatter(x=range(10),
            y=y_test.head(10),
            color="r",
           label="Actual Points")
plt.scatter(x=range(10),
            y=y_predict_test[:10],
            color="g",
           label="Predicted Points")
plt.legend()
```
